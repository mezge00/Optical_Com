```python
import matplotlib.pyplot as plt

def plot_data_rates(data_rates):
    """
    Plots the given data rates over time.
    
    Args:
    data_rates (list of float): The data rates to plot.
    """
    # Generate a list of indices for the x-axis
    x = list(range(len(data_rates)))
    
    # Plot the data rates
    plt.figure(figsize=(10, 5))
    plt.plot(x, data_rates, marker='o', linestyle='-', color='b', label='Data Rate')
    
    # Adding title and labels
    plt.title('Data Rates Over Time')
    plt.xlabel('Time (seconds)')
    plt.ylabel('Data Rate (bytes/second)')
    
    # Show the legend
    plt.legend()
    
    # Show the plot
    plt.grid(True)
    plt.show()
data_rates = [
    65.35, 50.27, 690.81, 2373.39, 1692.23, 2146.29, 2414.12, 2379.74,
    2487.36, 2369.93, 2395.23, 2395.86, 2452.87, 2401.17, 2176.79, 1867.58,
    2302.23, 2434.25, 2350.53, 1004.87, 688.84, 1375.78, 2309.85, 1716.51,
    2375.33, 2452.40, 2372.59, 2395.39, 870.86, 8.91, 1934.90, 1703.53,
    2437.57, 2369.06, 2452.19, 2286.26, 2357.75, 2346.14, 2420.28, 2398.74,
    2335.35, 2457.39, 2277.39, 2397.21, 863.28, 75.89, 1912.28, 2359.87,
    2423.66, 2396.85, 2429.17, 2362.27, 2382.52, 2298.18, 2392.47, 2456.56,
    2398.53, 2454.44, 2430.34, 2452.47, 2403.46, 2460.29, 2425.31, 2368.55,
    752.80
]

# Call the function to plot data rates
plot_data_rates(data_rates)
```
